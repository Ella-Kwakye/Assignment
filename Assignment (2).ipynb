```python
import numpy as np
import matplotlib.pyplot as plt

class Schrodinger:
    
    def __init__(self, a, b, V, n, hbar=1, m=1):
        self.a = a
        self.b = b
        self.V = V
        self.n = n
        self.hbar = hbar
        self.m = m
        self.dx = (b-a)/n
        self.x = np.linspace(a, b, n)
        self.psi = np.zeros(n)
        self.V_vec = np.zeros(n)
        for i in range(n):
            if a <= self.x[i] <= b:
                self.V_vec[i] = V
    
    def discretize(self):
        self.H = np.zeros((self.n,self.n))
        for i in range(1,self.n-1):
            self.H[i,i] = -2*self.hbar**2/(self.m*self.dx**2) + self.V_vec[i]
            self.H[i,i+1] = self.hbar**2/(self.m*self.dx**2)
            self.H[i,i-1] = self.hbar**2/(self.m*self.dx**2)
        self.H[0,0] = self.H[self.n-1,self.n-1] = 1
        
    def solve(self):
        self.e, self.psi = np.linalg.eigh(self.H)
        self.psi = self.psi/np.sqrt(self.dx)
    
    def plot_eigenvalues(self):
        plt.plot(self.e)
        plt.xlabel('Eigenvalue Index')
        plt.ylabel('Energy Eigenvalue')
        plt.show()
    
    def plot_wavefunction(self, index):
        plt.plot(self.x, self.psi[:,index])
        plt.xlabel('Position')
        plt.ylabel('Wavefunction Amplitude')
        plt.show()

a = 0
b = 1
V = 10
n = 1000
s = Schrodinger(a, b, V, n)
s.discretize()
s.solve()
s.plot_eigenvalues()
s.plot_wavefunction(0)
```
